# Tensor

## 산술연산

```python
a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])
c = torch.tensor([1, 1, 1])

print("[+] Addition")
print(f"a + b + c = {a + b + c}")
print(f"torch.add(a, b).add(c) = {torch.add(a, b).add(c)}")

print("\n[+] Subtraction")
print(f"a - b = {a - b}")
print(f"torch.sub(a, b) = {torch.sub(a, b)}")

print("\n[+] Element-wise Multiplication")
print(f"a * b = {a * b}")
print(f"torch.mul(a, b) = {torch.mul(a, b)}")

print("\n[+] Element-wise Division")
print(f"a / b = {a / b}")
print(f"torch.div(a, b) = {torch.div(a, b)}")
```

```python
# In-place operations (modifies the tensor)
d = torch.tensor([1, 2, 3])
print(f"[+] Original d = {d}")

d.add_(b)  # underscore suffix : in-place operations. B에 E를 더한다.
print(f"[+] After d.add_(b), d = {d}")
```

# PyTorch 학습 과정 개념

## 순전파(Forward)

```python
# 1) 예제 데이터 (x: 입력)
x = torch.tensor([[1.0],
                  [2.0],
                  [3.0]])  # (3, 1) : 샘플 3개(항목의 개수), 특징 1개 (항목 수)
```

```python
# 2) 가중치와 편향 (여기서는 그냥 수동으로 줌)
w = torch.tensor([[2.0]])  # (1, 1)
b = torch.tensor([1.0])    # (1,)
```

```python
# 3) 순전파: y_pred = x @ w + b
y_pred = x @ w + b  # 또는 torch.matmul(x, w) + b X에서 특징 1개와 W의 특징 1의 숫자가 같아야 한다.
```

```python
print("x shape:", x.shape)
print("y_pred shape:", y_pred.shape)
print("y_pred:\n", y_pred)
```

## 손실 계산(MSE Loss)

```python
# 1) 입력, 정답, 모델 파라미터
x = torch.tensor([[1.0],
                  [2.0],
                  [3.0]])
y_true = torch.tensor([[3.0],
                       [5.0],
                       [7.0]])  # 실제로 y = 2x + 1
```

```python
# 2) 가중치와 편향 (일부러 틀린 값)
w = torch.tensor([[1.5]]) #2가 원래는 정답값
b = torch.tensor([0.5]) #1이 원래는 정답값
```

```python
# 3) 순전파
y_pred = x @ w + b
```

```python
# 4) 손실 계산 (MSE)
loss = torch.mean((y_pred - y_true) ** 2) #**2 = 거듭제급, 음수를 양수로 변경하기 위해
```

```python
print("y_pred:\n", y_pred)
print("y_true:\n", y_true)
print("loss:", loss.item())
```

## 역전파(Backpropagation)

```python
# 1) 데이터
x = torch.tensor([[1.0],
                  [2.0],
                  [3.0]])

y_true = torch.tensor([[3.0],
                       [5.0],
                       [7.0]])
```

```python
# 2) 파라미터: requires_grad=True 로 설정
w = torch.tensor([[1.5]], requires_grad=True)
b = torch.tensor([0.5],  requires_grad=True)
```

```python
# 3) 순전파
y_pred = x @ w + b
loss = torch.mean((y_pred - y_true) ** 2) #평균값을 구해 초기 로스값을 설정
```

```python
print("초기 loss:", loss.item())
```

```python
# 4) 역전파
loss.backward()  # d(loss)/dw, d(loss)/db를 계산
```

```python
print("w.grad:", w.grad)
print("b.grad:", b.grad) # 손실을 줄이기 위해 w, b를 어느 방향으로/얼마나 바꿀지를 알려주는 정보
```

## 가중치 업데이트(Weight Update)

```python
x = torch.tensor([[1.0],
                  [2.0],
                  [3.0]])

y_true = torch.tensor([[3.0],
                       [5.0],
                       [7.0]])
```

```python
# 학습 가능한 파라미터
w = torch.tensor([[1.5]], requires_grad=True)
b = torch.tensor([0.5],  requires_grad=True)
```

```python
learning_rate = 0.1
```

```python
for step in range(20): #20번 계산
    # 1) 순전파
    y_pred = x @ w + b

    # 2) 손실계산
    loss = torch.mean((y_pred - y_true) ** 2)

    # 기존 gradient 초기화 : PyTorch는 backward() 호출시 gradient를 누적함. step별 gradient 쓰기 위한 방법.
    if w.grad is not None:
        w.grad.zero_() #w값을 초기화하는 설정, 이전 값이 영향을 받지 않기 위해
        b.grad.zero_() #b값을 초기화하는 설정, 이전 값이 영향을 받지 않기 위해

    # 3) 역전파
    loss.backward()

    # 4) 가중치 업데이트 (no_grad 블록 안에서)
    with torch.no_grad():
        w -= learning_rate * w.grad
        b -= learning_rate * b.grad

    print(f"step {step+1:2d} | loss = {loss.item():.4f} | w = {w.item():.4f}, b = {b.item():.4f}")
```

# 미니 학습 예제 : Optimizer(SGD/Adam) 사용

```python
x = torch.tensor([[1.0],
                  [2.0],
                  [3.0]])

y_true = torch.tensor([[3.0],
                       [5.0],
                       [7.0]])
```

```python
# 파라미터
w = torch.tensor([[1.5]], requires_grad=True)
b = torch.tensor([0.5],  requires_grad=True)
```

```python
# 파라미터를 묶어서 optimizer에 넘기기
optimizer = torch.optim.SGD([w, b], lr=0.1)
```

```python
for step in range(20):
    # 1) 순전파
    y_pred = x @ w + b
    loss = torch.mean((y_pred - y_true) ** 2)

    # 2) gradient 초기화
    optimizer.zero_grad()

    # 3) 역전파
    loss.backward()

    # 4) optimizer가 내부적으로 w, b 업데이트
    optimizer.step()

    print(f"step {step+1:2d} | loss = {loss.item():.4f} | w = {w.item():.4f}, b = {b.item():.4f}")
```
